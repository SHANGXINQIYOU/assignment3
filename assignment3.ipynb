```python
import random

def get_final_number():
    
    numbers = list(range(1, 2025))
    
    while len(numbers) > 1:
        
        num1, num2 = numbers.pop(random.randint(0,len(numbers)-1)),numbers.pop(random.randint(0,len(numbers)-1))
        
        difference = abs(num1 - num2)
                
        numbers.append(difference)

    return numbers[0]

results = [get_final_number() for _ in range(500)]
print(" The result is:", results)
```

```python
import matplotlib.pyplot as plt

def get_final_number():
    
    numbers = list(range(1, 2025))
    
    while len(numbers) > 1:
        
        num1, num2 = numbers.pop(random.randint(0,len(numbers)-1)),numbers.pop(random.randint(0,len(numbers)-1))
        
        difference = abs(num1 - num2)
                
        numbers.append(difference)

    return numbers[0]

results = [get_final_number() for _ in range(100000)]

# plt.show
# print(" The result is:", results)
```
